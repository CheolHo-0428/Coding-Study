### My Code

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

naver_music_url = 'https://music.naver.com/listen/history/index.nhn?type=TOTAL_V2&year=2020&month=10&week=0&page='

for page in range(2) :
    if page == 0 :
        naver_music_url_page = naver_music_url + str(page+1)
     
        html_naver_music = requests.get(naver_music_url_page).text
        soup_naver_music = BeautifulSoup(html_naver_music, 'lxml')

        naver_titles = soup_naver_music.select('a._title span.ellipsis')
        naver_artists = soup_naver_music.select('td._artist a')

        naver_music_titles50 = [naver_title.get_text() for naver_title in naver_titles]
        naver_music_artists50 = [naver_artist.get_text().strip() for naver_artist in naver_artists]
    
    else :
        naver_music_url_page = naver_music_url + str(page+1)
     
        html_naver_music = requests.get(naver_music_url_page).text
        soup_naver_music = BeautifulSoup(html_naver_music, 'lxml')

        naver_titles = soup_naver_music.select('a._title span.ellipsis')
        naver_artists = soup_naver_music.select('td._artist a')

        naver_music_titles100 = [naver_title.get_text() for naver_title in naver_titles]
        naver_music_artists100 = [naver_artist.get_text().strip() for naver_artist in naver_artists]

total_naver_music_titles = naver_music_titles50 + naver_music_titles100
total_naver_music_artists = naver_music_artists50 + naver_music_artists100

excel_naver_music_data = {'순위' : list(range(1,101)), 
                         'Title' : total_naver_music_titles , 
                        'Artist' : total_naver_music_artists}

df = pd.DataFrame(excel_naver_music_data, columns=['순위', 'Title', 'Artist'])

excel_writer = pd.ExcelWriter('c:/myPyCode/data/Naver_Music_Top100.xlsx', 
                              engine='xlsxwriter')
df.to_excel(excel_writer, index = False)
excel_writer.save()
```

### Another Code(추출한 네이버뮤직 TOP100을 엑셀파일에 저장하기)

```python
import requests
from bs4 import BeautifulSoup
import openpyxl

wb = openpyxl.Workbook()
sheet = wb.active
sheet.append(["순위", "노래 제목", "아티스트"])

naver_music_url = "http://music.naver.com/listen/history/index.nhn?type=DOMESTIC&year=2017&month=12&week=1&page="

# 네이버 music 주소를 입력하면 노래 제목과 아티스트를 반환
def naver_music(url):    
    html_music = requests.get(url).text
    soup_music = BeautifulSoup(html_music, "lxml")

    titles = soup_music.select('a._title span.ellipsis') 
    artists = soup_music.select('td._artist a')

    music_titles = [title.get_text() for title in titles]
    music_artists = [artist.get_text().strip() for artist in artists]
    
    return music_titles, music_artists

# 각 page에는 50개의 노래 제목과 아티스트가 추출됨
for page in range(2):
    naver_music_url_page = naver_music_url + str(page+1) # page URL
    naver_music_titles, naver_music_artists = naver_music(naver_music_url_page)
    
    for k in range(len(naver_music_titles)):
        sheet.append([page*50 + k+1,naver_music_titles[k], naver_music_artists[k]])
    
wb.save("NaverMusicTop100.xlsx")
```
