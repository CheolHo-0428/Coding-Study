```python
import pandas as pd
from bs4 import BeautifulSoup
import glob
from selenium import webdriver
import requests 
import os
import time
import numpy as np
from matplotlib import pyplot as plt
import matplotlib

#    
receipt_data_file = glob.glob('C:/Users/Playdata/Desktop/shinhancard/shinhancard_*.html') #       

# excel_writer  data 
total_df = []             #  DataFrame 
total_df_region = []      #  DataFrame 
List_Sum_Month_tdf = []   #   DataFrame 
sheet_nm = []             # Excel Sheet Name 
visit_graph = []          #    

#  'total_df' 'sheet_nm'    for 
for url in receipt_data_file :
    driver = webdriver.Chrome('c:/chromedriver.exe')
    driver.get(url)
    html = driver.page_source
    soup = BeautifulSoup(html,'html.parser')

    title_tag = soup.select('head > title')
    title = title_tag[0].text
    
    #    if  
    if title == '' :
        pw = '950730'
        driver.find_element_by_id('password').send_keys(pw)
        driver.find_element_by_xpath('//*[@id="aform"]/span/input').click()
        time.sleep(5)
        html = driver.page_source
        soup = BeautifulSoup(html,'html.parser')
        
    #       
    #      
    tables = soup.select('div#email02 > table.tblList.cardUseList > tbody > tr')
    tables_middleSum = soup.select('div#email02 > table.tblList.cardUseList > tbody > tr.middleSum')
    
    for n in range(len(tables_middleSum)) :
        tables.remove(tables_middleSum[n])
    
    # tables  string                        
    seperate_tables = []                        # seperate_tables 
    for n in range(len(tables)) :
        table = str(tables[n].text).split('\n')
        seperate_tables = seperate_tables + table
        
    # seperate_tables '' , '', ''    
    date_list =  [seperate_tables[n] for n in range(1,len(seperate_tables), 12)]
    price_list = [int(seperate_tables[n].replace(',','')) for n in range(4,len(seperate_tables), 12)] 
    store_list = [seperate_tables[n].replace(" ", "") for n in range(3,len(seperate_tables), 12)]

    # seperate_tables  ''  -->   API --> ''   
    location_list = []
    def create_url(search_text):
        headers = {'Authorization': 'KakaoAK 3e8ea332f0cdef6293383ac7bf81a6fc'}
        base = "https://dapi.kakao.com/v2/local/search/keyword.json"
        param_query = "?query=" +search_text
        respond = requests.get(base + param_query,headers=headers)
        return respond.json()

    for n in range(len(store_list)) :
        store_name = store_list[n]
        result = create_url(store_name)
        tmp1 = result['documents']
        try :
            tmp2 = tmp1[0]['address_name']
            location_list.append(tmp2)
        except :
            tmp2 = np.nan
            location_list.append(tmp2)
            
    #   DataFrame 
    table_data = {'':date_list, '':store_list, '':location_list,  '':price_list }
    
    df = pd.DataFrame(table_data, index = range(1,len(date_list)+1,1), columns=['', '', '', ''])
    df_sum = pd.DataFrame(df.sum(), columns=[''])
    df_total = df.append(df_sum.T)
    df_total.loc['', ''] = ''
    df_total.loc['', ''] = ''
    df_total.loc['', ''] = ''
    total_df.append(df_total)
    List_Sum_Month_tdf.append(df_sum.T[''])    #     List_Sum_Month_tdf=[] .
        
   #      
    df_dis_nan = df.dropna()

    df_address_list = list(df_dis_nan[''])
    df_region_list = [ ' '.join(address.split()[:2]) for address in df_address_list]

    df_price_list = list(df_dis_nan[''])

    table_data_1 = {'':df_region_list, '':df_price_list}
    col_list = ['', '']
    df_Dup = pd.DataFrame(table_data_1 , columns = col_list ) #   df
    df_NoDup = df_Dup.drop_duplicates(subset = '')  #   df

    index_name_1 = df_Dup['']  #    series
    df_Dup_name = df_Dup.rename(index_name_1)
    df_Dup_name = df_Dup_name.drop([''], axis = 1)  #    index df

    index_name_2 = df_NoDup['']  #    series
    df_NoDup_name = df_NoDup.rename(index_name_2)
    df_NoDup_name = df_NoDup_name.drop([''], axis = 1)  #    index df

    for nm in list(index_name_2) :
        price_data = df_Dup_name.loc[nm]['']
        price_sum = price_data.sum()
        df_NoDup_name.loc[nm][''] = price_sum

    name_region = df_Dup['']
    visit_counter = dict(name_region.value_counts())
    counter_list = []
    for k in index_name_2 :
        counter_list.append(visit_counter[k])
    
    df_NoDup_name.insert(0, '', counter_list)
    total_df_region.append(df_NoDup_name)
    
    #    
    tables_date = soup.select('body > div.wrap.check > div.content > div.receiveDate > span.date')
    nm = str(tables_date[0].text)
    sheet_nm.append(nm)
    
    #  
    driver.close()

#      
folder = 'C:/Users/Playdata/Desktop/ '
folder_1 = 'C:/Users/Playdata/Desktop/ /image'

if not os.path.exists(folder) :
    os.makedirs(folder)

if not os.path.exists(folder_1) :
    os.makedirs(folder_1)
    
#    (   )
     
mean_sales = int(np.mean(List_Sum_Month_tdf))

fig = plt.figure(figsize=(5.5,5.5)) ##  
fig.set_facecolor('white') ##   
ax = fig.add_subplot() ##  () 

ax.spines['right'].set_visible(False) ##   
ax.spines['top'].set_visible(False) ##   

args_plot = [sheet_nm, List_Sum_Month_tdf] ##  
config_plot = dict( ##  
    color='red', #  
    linestyle='solid', #  
    linewidth=1, #   
    marker='o', #  
    markersize=5, #  
    markeredgecolor='white', #   
    markeredgewidth=1.5, #   
)

line_plot = ax.plot(label='', *args_plot,**config_plot) ##   -  
line_plot = line_plot[0]

ax.axhline(mean_sales,label=' ') ##  y   
ax.text(1,mean_sales+(15000/2),f'  : {mean_sales}',fontsize=7.5) ##    

bbox = dict( boxstyle='square', facecolor='white')

for coord in list(line_plot.get_xydata()):
    ax.text(coord[0],coord[1]+20000,f'{int(coord[1])}',fontsize=6, ha='center', bbox=bbox)

ylim = ax.get_ylim() ##  y  

yticks = list(ax.get_yticks()) ## y  .

for y in yticks:
    ax.axhline(y,linestyle=(0,(5,2)),color='grey',alpha=0.5) ##  
    
yticks.append(mean_sales)
yticks = sorted(yticks)

ax.set_yticks(yticks) ##   y  .
ax.set_ylim(ylim) ##  y  
ax.legend(loc='upper left', fontsize=7.5, shadow=True) ##   -   

plt.gca().get_yaxis().get_major_formatter().set_scientific(False) # Y  
plt.yticks(fontsize=6.5) ## y   
plt.xticks(rotation=360,fontsize=6) ## x    

plt.ylabel('',rotation=90, fontsize=10, color='blue') ## y 
plt.xlabel('',fontsize=10, color='blue') ## x 

plt.title('2019 ~ 2020  ',fontsize=10, color='blue') ##  

graph_file = "C:/Users/Playdata/Desktop/ /image/mcua.png"

plt.savefig(graph_file, dpi = 200)
    
#      

for n in range(len(total_df_region)) :
    a0 = total_df_region[n][[""]].plot.bar(grid = True, rot=45, figsize=[10,8], fontsize = 15)
    a0.set_xlabel("",fontsize = 15)
    a0.set_ylabel(" ",fontsize = 15)
    a0.set_title(f"{sheet_nm[n]}  ", fontsize = 20)
    plt.tick_params(axis='x', labelsize=10)
    plt.tick_params(axis='y', labelsize=10)
    plt.xticks(rotation=45)
    visit_file = f"C:/Users/Playdata/Desktop/ /image/{sheet_nm[n]}   .png" 
    plt.savefig(visit_file,dpi = 200)
    visit_graph.append(visit_file)

# total_df, total_df_region, sheet_nm   Excel 
    
excel_writer = pd.ExcelWriter('C:/Users/Playdata/Desktop/ /CardUseList.xlsx', engine='xlsxwriter')

df_empty = pd.DataFrame()
df_empty.to_excel(excel_writer, index = True, sheet_name = 'Sheet1')
worksheet = excel_writer.sheets['Sheet1']
worksheet.insert_image('D2', graph_file)

ToExcel_data =[]

for n in range(len(total_df)) :
    ToAppend_total_df = total_df[n].to_excel(excel_writer, index = True, sheet_name = sheet_nm[n]+' ')
    ToExcel_data.append(ToAppend_total_df)
    
    ToAppend_total_region_df = total_df_region[n].to_excel(excel_writer, index = True, sheet_name = sheet_nm[n]+' ', startcol=6)
    worksheet_1 = excel_writer.sheets[sheet_nm[n]+' ']
    send_worksheet_1 = worksheet_1.insert_image('K1', visit_graph[n])
    ToExcel_data.append(ToAppend_total_region_df)
    ToExcel_data.append(send_worksheet_1)

excel_writer.save()
```
