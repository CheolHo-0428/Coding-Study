```python
import pandas as pd
from bs4 import BeautifulSoup
import glob
from selenium import webdriver
import requests 
import os
import time
import numpy as np
from matplotlib import pyplot as plt
import matplotlib
import seaborn as sns

matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False 

#    
receipt_data_file = glob.glob('C:/Users/Playdata/Desktop/shinhancard/shinhancard_*.html') #       

# excel_writer  data 
total_df = []             #  DataFrame 
total_df_region = []      #  DataFrame 
List_Sum_Month_tdf = []   #   DataFrame 
sheet_nm = []             # Excel Sheet Name 
visit_graph = []          #     
price_graph = []         #     


#  'total_df' 'sheet_nm'    for 
for url in receipt_data_file :
    driver = webdriver.Chrome('c:/chromedriver.exe')
    driver.get(url)
    html = driver.page_source
    soup = BeautifulSoup(html,'html.parser')

    title_tag = soup.select('head > title')
    title = title_tag[0].text
    
    #    if  
    if title == '' :
        pw = '950730'
        driver.find_element_by_id('password').send_keys(pw)
        driver.find_element_by_xpath('//*[@id="aform"]/span/input').click()
        time.sleep(5)
        html = driver.page_source
        soup = BeautifulSoup(html,'html.parser')
        
    #       
    #      
    tables = soup.select('div#email02 > table.tblList.cardUseList > tbody > tr')
    tables_middleSum = soup.select('div#email02 > table.tblList.cardUseList > tbody > tr.middleSum')
    
    for n in range(len(tables_middleSum)) :
        tables.remove(tables_middleSum[n])
    
    # tables  string                        
    seperate_tables = []                        # seperate_tables 
    for n in range(len(tables)) :
        table = str(tables[n].text).split('\n')
        seperate_tables = seperate_tables + table
        
    # seperate_tables '' , '', ''    
    date_list =  [seperate_tables[n] for n in range(1,len(seperate_tables), 12)]
    price_list = [int(seperate_tables[n].replace(',','')) for n in range(4,len(seperate_tables), 12)] 
    store_list = [seperate_tables[n].replace(" ", "") for n in range(3,len(seperate_tables), 12)]

    # seperate_tables  ''  -->   API --> ''   
    location_list = []
    def create_url(search_text):
        headers = {'Authorization': 'KakaoAK 3e8ea332f0cdef6293383ac7bf81a6fc'}
        base = "https://dapi.kakao.com/v2/local/search/keyword.json"
        param_query = "?query=" +search_text
        respond = requests.get(base + param_query,headers=headers)
        return respond.json()

    for n in range(len(store_list)) :
        store_name = store_list[n]
        result = create_url(store_name)
        tmp1 = result['documents']
        try :
            tmp2 = tmp1[0]['address_name']
            location_list.append(tmp2)
        except :
            tmp2 = np.nan
            location_list.append(tmp2)
            
    #   DataFrame 
    table_data = {'':date_list, '':store_list, '':location_list,  '':price_list }
    
    df = pd.DataFrame(table_data, index = range(1,len(date_list)+1,1), columns=['', '', '', ''])
    df_sum = pd.DataFrame(df.sum(), columns=[''])
    df_total = df.append(df_sum.T)
    df_total.loc['', ''] = ''
    df_total.loc['', ''] = ''
    df_total.loc['', ''] = ''
    total_df.append(df_total)
    List_Sum_Month_tdf.append(df_sum.T[''])    #     List_Sum_Month_tdf=[] .
        
   #      
    df_dis_nan = df.dropna()

    df_address_list = list(df_dis_nan[''])
    df_region_list = [ ' '.join(address.split()[:2]) for address in df_address_list]

    df_price_list = list(df_dis_nan[''])

    table_data_1 = {'':df_region_list, '':df_price_list}
    col_list = ['', '']
    df_Dup = pd.DataFrame(table_data_1 , columns = col_list ) #   df
    df_NoDup = df_Dup.drop_duplicates(subset = '')  #   df

    index_name_1 = df_Dup['']  #    series
    df_Dup_name = df_Dup.rename(index_name_1)
    df_Dup_name = df_Dup_name.drop([''], axis = 1)  #    index df

    index_name_2 = df_NoDup['']  #    series
    df_NoDup_name = df_NoDup.rename(index_name_2)
    df_NoDup_name = df_NoDup_name.drop([''], axis = 1)  #    index df

    for nm in list(index_name_2) :
        price_data = df_Dup_name.loc[nm]['']
        price_sum = price_data.sum()
        df_NoDup_name.loc[nm][''] = price_sum

    name_region = df_Dup['']
    visit_counter = dict(name_region.value_counts())
    counter_list = []
    for k in index_name_2 :
        counter_list.append(visit_counter[k])
    
    df_NoDup_name.insert(0, '', counter_list)
    total_df_region.append(df_NoDup_name)
    
    #    
    tables_date = soup.select('body > div.wrap.check > div.content > div.receiveDate > span.date')
    nm = str(tables_date[0].text)
    sheet_nm.append(nm)
    
    #  
    driver.close()

#      
folder = 'C:/Users/Playdata/Desktop/ '
folder_1 = 'C:/Users/Playdata/Desktop/ /image'

if not os.path.exists(folder) :
    os.makedirs(folder)

if not os.path.exists(folder_1) :
    os.makedirs(folder_1)
    
#    (   )
     
mean_sales = int(np.mean(List_Sum_Month_tdf))

fig = plt.figure(figsize=(8,8)) ##  
fig.set_facecolor('white') ##   
ax = fig.add_subplot() ##  () 

ax.spines['right'].set_visible(False) ##   
ax.spines['top'].set_visible(False) ##   

args_plot = [sheet_nm, List_Sum_Month_tdf] ##  
config_plot = dict( ##  
    color='red', #  
    linestyle='solid', #  
    linewidth=2, #   
    marker='o', #  
    markersize=10, #  
    markeredgecolor='white', #   
    markeredgewidth=3, #   
)

line_plot = ax.plot(label='', *args_plot,**config_plot) ##   -  
line_plot = line_plot[0]

ax.axhline(mean_sales,label=' ') ##  y   
ax.text(2,mean_sales+15000,f'  : {mean_sales}',fontsize=12) ##    

bbox = dict( boxstyle='square', facecolor='white')

for coord in list(line_plot.get_xydata()):
    ax.text(coord[0],coord[1]+20000,f'{int(coord[1])}',fontsize=9.6, ha='center', bbox=bbox)

ylim = ax.get_ylim() ##  y  

yticks = list(ax.get_yticks()) ## y  .

for y in yticks:
    ax.axhline(y,linestyle=(0,(5,2)),color='grey',alpha=0.5) ##  
    
yticks.append(mean_sales)
yticks = sorted(yticks)

ax.set_yticks(yticks) ##   y  .
ax.set_ylim(ylim) ##  y  
ax.legend(loc='upper left', fontsize=12, shadow=True) ##   -   

plt.gca().get_yaxis().get_major_formatter().set_scientific(False) # Y  
plt.yticks(fontsize=10.4) ## y   
plt.xticks(rotation=360,fontsize=9.6) ## x    

plt.ylabel('',rotation=90, fontsize=16, color='black') ## y 
plt.xlabel('',fontsize=10.4, color='black') ## x 

plt.title('2019 ~ 2020  ',fontsize=16, color='black') ##  

graph_file = "C:/Users/Playdata/Desktop/ /image/mcua.png"

plt.savefig(graph_file, dpi = 200)

##########   dataframe #############################

df_region_Dup = pd.DataFrame()        #   df_region

for n in range(len(total_df_region)) :
        df_region_Dup = df_region_Dup.append(total_df_region[n])

sorted_df_region_Dup = df_region_Dup.sort_index(ascending=True)  #     df_region

index_region = sorted_df_region_Dup.reset_index().rename(columns={'index': ''}) # index columns 

df_region_NoDup = index_region.drop_duplicates(subset = '')  #   df_region

index_name_3 = df_region_NoDup['']  
df_region_NoDup_name = df_region_NoDup.rename(index_name_3)
df_region_NoDup_name = df_region_NoDup_name.drop([''], axis = 1) #    index df_region

for nm_1 in list(index_name_3) :
    price_region_data = sorted_df_region_Dup.loc[nm_1]['']
    price_region_sum = price_region_data.sum()
    df_region_NoDup_name.loc[nm_1][''] = price_region_sum        #     
    
    visit_region_data = sorted_df_region_Dup.loc[nm_1]['']
    visit_region_sum = visit_region_data.sum()
    df_region_NoDup_name.loc[nm_1][''] = visit_region_sum        #     

################ 1     ################
## 
total_region_category = index_name_3
total_visit_category = df_region_NoDup_name['']
total_price_category = df_region_NoDup_name['']

## 
fig = plt.figure(figsize=(8,8)) ##  
fig.set_facecolor('white') ##   
ax = fig.add_subplot() ##  () 
ax.spines['right'].set_visible(False) ##   
ax.spines['top'].set_visible(False) ##   

ytick_label_position = list(range(len(total_region_category))) ## y    x
plt.yticks(ytick_label_position, total_region_category) ## y   

plt.barh(ytick_label_position, total_visit_category, color = 'cadetblue') ##  

for i, v in enumerate(total_visit_category):
    plt.text(v + .5, i - .3, str(v), color='black', fontweight='bold')
    
plt.title('2019~2020  ',fontsize=16) ##  
plt.xlabel('',fontsize=10.4) ## x  
plt.ylabel('',fontsize=8) ## y  

graph_file_1 = "C:/Users/Playdata/Desktop/ /image/total_visit.png"
plt.savefig(graph_file_1, dpi = 200)
plt.show()

##### 2019~2020    ####
fig = plt.figure(figsize=(8,8)) ##  
fig.set_facecolor('white') ##   
ax = fig.add_subplot() ##  () 
ax.spines['right'].set_visible(False) ##   
ax.spines['top'].set_visible(False) ##   

ytick_label_position_1 = list(range(len(total_region_category))) ## y    x
plt.yticks(ytick_label_position_1, total_region_category) ## y   
plt.gca().get_xaxis().get_major_formatter().set_scientific(False) # Y  

plt.barh(ytick_label_position_1, total_price_category, color = 'salmon') ##  

for i, v in enumerate(total_price_category):
    plt.text(v + 1, i - .3, str(v), color='black', fontweight='bold')
    
plt.title('2019~2020  ',fontsize=16) ##  
plt.xlabel('',fontsize=10.4) ## x  
plt.ylabel('',fontsize=8) ## y  

graph_file_2 = "C:/Users/Playdata/Desktop/ /image/total_price.png"
plt.savefig(graph_file_2, dpi = 200)

plt.show()
###################################################

#       (    )

for n in range(len(total_df_region)) :
    c_name = ['', '']
    ta = total_df_region[n]
    sn = sheet_nm[n]
    for k in range(len(c_name)) :
            region_category = ta.index
            visit_category = ta[c_name[k]]
            
            tick_size_1 = 6.5 ##   
            axis_label_size_1 = 7.5 ## x, y  

            fig_1 = plt.figure(figsize=(5,5)) ## Figure   10 by 10
            ax_1 = fig_1.add_subplot() ## Axes 
            colors_1 = sns.color_palette('hls',len(region_category)) ##  

            xtick_label_position = list(range(len(region_category))) ## x    x

            plt.yticks(fontsize=tick_size_1)
            plt.xticks(xtick_label_position, region_category, rotation = 30, fontsize=tick_size_1) ## x   

            bars = plt.bar(xtick_label_position, visit_category, color=colors_1, \
                           edgecolor=colors_1, alpha=0.35, linewidth=1.5) ##  

            for i, b in enumerate(bars):
                ax_1.text(b.get_x()+b.get_width()*(1/2),b.get_height()+0.1, \
                        visit_category[i],ha='center',fontsize=6.5)

            plt.title(f'{sn}  {c_name[k]}',fontsize=10) ##  
            plt.xlabel('',fontsize=axis_label_size_1) ## x  
            plt.ylabel(c_name[k],fontsize=axis_label_size_1) ## y  
            visit_file = f"C:/Users/Playdata/Desktop/ /image/{sn}_{c_name[k]}.png"
            plt.savefig(visit_file, dpi = 200)
            
            if c_name[k] == '' : 
                visit_graph.append(visit_file)
                
            else :
                price_graph.append(visit_file)
            
# total_df, total_df_region, sheet_nm   Excel 
    
excel_writer = pd.ExcelWriter('C:/Users/Playdata/Desktop/ /CardUseList.xlsx', engine='xlsxwriter')

df_empty = pd.DataFrame()
df_empty.to_excel(excel_writer, index = True, sheet_name = 'Sheet1')
worksheet = excel_writer.sheets['Sheet1']
worksheet.insert_image('A1', graph_file) 
worksheet.insert_image('N1', graph_file_1)
worksheet.insert_image('AA1', graph_file_2)


ToExcel_data =[]

for n in range(len(total_df)) :
    ToAppend_total_df = total_df[n].to_excel(excel_writer, index = True, sheet_name = sheet_nm[n]+' ')
    ToExcel_data.append(ToAppend_total_df)
    
    ToAppend_total_region_df = total_df_region[n].to_excel(excel_writer, index = True, sheet_name = sheet_nm[n]+' ', startcol=6)
    worksheet_1 = excel_writer.sheets[sheet_nm[n]+' ']
    send_worksheet_1 = worksheet_1.insert_image('K1', visit_graph[n])
    send_worksheet_2 = worksheet_1.insert_image('S1', price_graph[n])
    ToExcel_data.append(ToAppend_total_region_df)
    ToExcel_data.append(send_worksheet_1)

excel_writer.save()
```

```python
df_address = [] #     DF  
df_address_only = [] #    DF  
data_in_name = []
data_in_only_name = []
count_list = []

#     DF 
df_address = [df.dropna() for df in total_df]

#    DF 
df_address_only = [df.drop_duplicates(subset = '') for df in df_address]

for i in range(len(df_address)):
    
    name_index =  df_address[i]['']
    data_in_name.append(df_address[i].rename(name_index))
    data_in_name[i] = data_in_name[i].drop([''], axis = 1)

    name_index =  df_address_only[i]['']
    data_in_only_name.append(df_address_only[i].rename(name_index))
    data_in_only_name[i] = data_in_only_name[i].drop([''], axis = 1)

    name_series = df_address[i]['']

    visit_count = dict(name_series.value_counts())

    count_list_tmp = []

    for k in name_index:
        count_list_tmp.append(visit_count[k])

        
    count_list.append(count_list_tmp)
    
    #   

    data_in_only_name[i].insert(3, '', count_list[i])

import requests

coordinates_x = []
coordinates_y = []

def getLatLng(address):
    result = ""
 
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
    rest_api_key = '289858dc68417f30677b064987b71bab'
    header = {'Authorization': 'KakaoAK ' + rest_api_key}
 
    r = requests.get(url, headers=header)
 
    if r.status_code == 200:
        try:
            result_address = r.json()["documents"][0]["address"]
            result = result_address["y"], result_address["x"]
        except:
            result = False
    else:
        result = "ERROR[" + str(r.status_code) + "]"
    
    return result

for i in range(len(data_in_only_name)):

    date = list(data_in_only_name[i][''][:-1])
    store_name = list(data_in_only_name[i].index[:-1])
    address = list(data_in_only_name[i][''][:-1])
    price = list(data_in_only_name[i][''][:-1])
    count = count_list[i][:-1]

    coordinates_x = []
    coordinates_y = []
    
    for j in address:
        
        coor = getLatLng(j)
        

        if coor != False:
            coordinates_x.append(coor[0])
            coordinates_y.append(coor[1])
        else:
            coordinates_x.append(False)
            coordinates_y.append(False)

    html = '<div id="map" style="width:100%;height:100%;"></div>'
    html = html + '<script type="text/javascript" src="//dapi.kakao.com/v2/maps/sdk.js?appkey=562beaefa8f3396ba21741328654bece"></script>\n'
    html = html + '<script>\n'
    html = html + "var mapContainer = document.getElementById('map'), \n"
    html = html + '    mapOption = { \n'

    html = html + '        center: new kakao.maps.LatLng(37.566826, 126.9786567), \n'
    html = html + '        level: 7 \n'
    html = html + '    };'

    html = html + """
    var map = new kakao.maps.Map(mapContainer, mapOption); //  

    //    title   

    var positions = []
    """
    for k in range(len(store_name)):
        if coordinates_x[k] == False:
            continue
        html = html + "positions.push({ title: " + f"'{store_name[k]}', price: '{price[k]}', count: '{count[i]}'," + f" latlng: new kakao.maps.LatLng({coordinates_x[k]}, {coordinates_y[k]})" + "})\n"

    html = html + """
    //    
    var imageSrc = "https://t1.daumcdn.net/localimg/localimages/07/mapapidoc/markerStar.png"; 

    for (var i = 0; i < positions.length; i ++) {

        //     
        var imageSize = new kakao.maps.Size(24, 35); 

        //       
        var markerImage = new kakao.maps.MarkerImage(imageSrc, imageSize); 

        //  
        var marker = new kakao.maps.Marker({
            map: map, //   
            position: positions[i].latlng, //   
            title : positions[i].title, //  ,     
            image : markerImage //   
        });

        var infowindow = new kakao.maps.InfoWindow({
            position : positions[i].latlng, 
            content : '<div style="width:100%;height:100%;font-size:11px"> : ' + positions[i].title + '<br> : ' + positions[i].count + '<br> : ' + positions[i].price +  '</div>'
        });


        infowindow.open(map, marker); 

    }
    </script>"""

    map_file = open('map_' + sheet_nm[i] + '.html', 'w')
    map_file.write(html)
    map_file.close()

    print(html)
```
